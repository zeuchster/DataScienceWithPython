```python
import numpy as np
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

## Adjusting the plots

```python
plt.plot(x,np.sin(x),color="blue")
plt.plot(x,np.sin(x+1),color="g")
plt.plot(x,np.sin(x+2),color="0.8")
plt.plot(x,np.sin(x+3),color="#FFDD44")
plt.show()
```

```python
plt.plot(x,np.sin(x),color="blue",linestyle="solid")
plt.plot(x,np.sin(x+1),color="g",linestyle="dashed")
plt.plot(x,np.sin(x+2),color="0.8",linestyle="dashdot")
plt.plot(x,np.sin(x+3),color="#FFDD44",linestyle="dotted")
plt.show()
```

```python
plt.plot(x,np.sin(x),color="blue",linestyle="-")
plt.plot(x,np.sin(x+1),color="g",linestyle="--")
plt.plot(x,np.sin(x+2),color="0.8",linestyle="-.")
plt.plot(x,np.sin(x+3),color="#FFDD44",linestyle=":")
plt.show()
```

```python
plt.plot(x,np.sin(x),"-g")
plt.plot(x,np.sin(x+1),"--c")
plt.plot(x,np.sin(x+2),"-.k")
plt.plot(x,np.sin(x+3),":r")
plt.show()
```

```python
plt.plot(x,np.sin(x),"-g")
plt.xlim(-1,12)
plt.ylim(-5,5)
plt.show()
```

```python
plt.plot(x,np.sin(x),"-g")
plt.xlim(-1,12)
plt.ylim(-5,5)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("This is a title of a plot")
plt.show()
```

```python
plt.plot(x,np.sin(x),label="Sin x")
plt.plot(x,np.sin(x+1),label = "Sin(x+1)")
plt.legend()
plt.show()
```

```python
import pandas as pd
sp_x = [2, 3, 4, 6, 2, 8, 6, 9, 12, 1, 7, 1]
sp_y = [10, 3, 4, 12, 5, 12, 4, 5, 6, 8, 9, 10]
df = pd.DataFrame({"A":sp_x, "B":sp_y})
df.plot(kind="scatter", x = "A", y = "B")
plt.show()
```
